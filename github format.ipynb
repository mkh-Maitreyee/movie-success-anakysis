## Data exploration

### Subtask:
Explore the dataset to understand its structure, data types, missing values, and distributions of key variables.


**Reasoning**:
Examine the shape, info, descriptive statistics, and value counts of key variables to understand the dataset's structure and characteristics.



```python
# Examine the shape of the DataFrame.
print("Shape of the DataFrame:", df.shape)

# Get a summary of the DataFrame, including data types and non-null values.
print("\nInfo:")
df.info()

# Calculate descriptive statistics for numerical columns.
print("\nDescriptive Statistics:")
numerical_cols = ['budget', 'revenue', 'popularity', 'vote_average', 'runtime', 'vote_count']
display(df[numerical_cols].describe())

# Analyze the distribution of categorical variables.
print("\nValue Counts for 'original_language':")
display(df['original_language'].value_counts())
print("\nValue Counts for 'status':")
display(df['status'].value_counts())

# Investigate the correlation between numerical variables.
print("\nCorrelation Matrix:")
display(df[numerical_cols].corr())
```

## Data analysis

### Subtask:
Analyze the relationship between budget and revenue, popularity and vote average.  Also, analyze the frequency distribution of genres and original language.


**Reasoning**:
Calculate the correlation between 'budget' and 'revenue', and 'popularity' and 'vote_average'. Then analyze the frequency distribution of 'genres' and 'original_language'.



```python
import json

# Calculate the correlation between 'budget' and 'revenue'.
budget_revenue_correlation = df['budget'].corr(df['revenue'])
print(f"Correlation between budget and revenue: {budget_revenue_correlation}")

# Calculate the correlation between 'popularity' and 'vote_average'.
popularity_vote_correlation = df['popularity'].corr(df['vote_average'])
print(f"Correlation between popularity and vote average: {popularity_vote_correlation}")


# Analyze the frequency distribution of genres.
genre_counts = {}
for index, row in df.iterrows():
    genres_list = json.loads(row['genres'])
    for genre_dict in genres_list:
        genre_name = genre_dict['name']
        genre_counts[genre_name] = genre_counts.get(genre_name, 0) + 1

print("\nGenre Frequency Distribution:")
sorted_genre_counts = dict(sorted(genre_counts.items(), key=lambda item: item[1], reverse=True))
for genre, count in sorted_genre_counts.items():
    print(f"{genre}: {count}")

# Analyze the frequency distribution of original language.
language_counts = df['original_language'].value_counts()
print("\nOriginal Language Frequency Distribution:")
print(language_counts)
```

## Data visualization

### Subtask:
Visualize the distributions of key numerical variables, the relationship between budget and revenue, and the top genres or languages.


**Reasoning**:
Visualize the distributions of key numerical variables, the relationship between budget and revenue, and the top genres or languages using histograms, scatter plots, and bar charts.



```python
import matplotlib.pyplot as plt
import json

# 1. Histograms for numerical variables
plt.figure(figsize=(16, 12))

plt.subplot(2, 3, 1)
plt.hist(df['budget'], bins=30, log=True, color='skyblue')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.title('Distribution of Budget (Log Scale)')

plt.subplot(2, 3, 2)
plt.hist(df['revenue'], bins=30, log=True, color='lightcoral')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.title('Distribution of Revenue (Log Scale)')

plt.subplot(2, 3, 3)
plt.hist(df['popularity'], bins=30, color='lightgreen')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity')

plt.subplot(2, 3, 4)
plt.hist(df['vote_average'], bins=10, color='gold')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.title('Distribution of Vote Average')

plt.subplot(2, 3, 5)
plt.hist(df['runtime'], bins=30, color='plum')
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.title('Distribution of Runtime')

plt.tight_layout()
plt.show()


# 2. Scatter plot of budget vs. revenue
plt.figure(figsize=(8, 6))
plt.scatter(df['budget'], df['revenue'], alpha=0.5, color='teal')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs. Revenue')
plt.show()


# 3. Bar chart of top 10 genres
genre_counts = {}
for index, row in df.iterrows():
    genres_list = json.loads(row['genres'])
    for genre_dict in genres_list:
        genre_name = genre_dict['name']
        genre_counts[genre_name] = genre_counts.get(genre_name, 0) + 1

sorted_genre_counts = dict(sorted(genre_counts.items(), key=lambda item: item[1], reverse=True))
top_10_genres = dict(list(sorted_genre_counts.items())[:10])


plt.figure(figsize=(10, 6))
plt.bar(top_10_genres.keys(), top_10_genres.values(), color='orange')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Genres')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 4. Horizontal bar chart of top 10 original languages
top_10_languages = df['original_language'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_languages.index, top_10_languages.values, color='purple')
plt.xlabel('Frequency')
plt.ylabel('Original Language')
plt.title('Top 10 Most Frequent Original Languages')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
df['genres'] = df['genres'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
```

```python
from collections import Counter
genre_list = sum(df['genres'], [])
genre_count = Counter(genre_list)
```

```python
genres_df = pd.DataFrame(genre_count.items(), columns=['Genre', 'Count'])
genres_df = genres_df.sort_values('Count', ascending=False)
```

```python
sns.barplot(data=genres_df.head(10), x='Count', y='Genre')
plt.title("Top 10 Most Common Genres")
plt.show()
```

```python
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
df['profit'] = df['revenue'] - df['budget']
df_exploded = df.explode('genres')
profit_by_genre = df_exploded.groupby('genres')['profit'].mean().sort_values(ascending=False)

profit_by_genre.head(10).plot(kind='barh')
plt.title('Average Profit by Genre')
plt.xlabel('Average Profit')
plt.show()
```

# 🎬 Movie Success Analysis

## 📁 Dataset
TMDb 5000 Movies Dataset from [Kaggle](https://www.kaggle.com/datasets/tmdb/tmdb-movie-metadata)

## 🎯 Objective
Analyze what factors contribute to a movie’s success using budget, genre, runtime, ratings, and more.

---

## 📊 Key Insights

### 💰 Budget vs Revenue
- Higher-budget movies usually generate more revenue, but the relationship is not guaranteed.
- Some low-budget movies earned large profits, showing budget is not the only success factor.

### 🎭 Genre Profitability
- Most profitable genres on average:
  - 🎨 Animation
  - 🗺️ Adventure
  - 🧝 Fantasy
  - 🚀 Science Fiction
- Drama and Comedy are the most frequent but not always the most profitable.

### 🕒 Runtime and Ratings
- Movies between **90–120 minutes** tend to receive better ratings.
- Very short or very long movies are often rated lower.

### 🌟 Ratings vs Revenue
- Highly rated movies do **not always** make the most money.
- Many blockbusters have average ratings but generate high revenue, likely due to strong marketing or franchises.

### 📈 Release Trends
- Movie production rose steadily from 2000 to 2015.
- Many top-rated and highest-grossing films were released after 2010.

---

## ✅ Conclusion

Movie success is influenced by a mix of **budget, genre, runtime, and release timing**. While high budgets and certain genres increase chances of revenue success, factors like audience engagement, timing, and marketing also play a major role. The sweet spot for runtime lies around **100 minutes**, and genres like **Animation and Adventure** offer strong returns.
